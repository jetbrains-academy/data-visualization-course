```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

root_path = Path().cwd().parent.parent
data_path = root_path / "common" / "resources" / "datasets"
figures_path = Path().resolve() / "figures"
```

```python
def prepare_dataset(dataset: pd.DataFrame) -> pd.DataFrame:
    dataset = dataset[dataset.User_Score != "tbd"]
    dataset["User_Score"] = dataset["User_Score"].astype("float")
    dataset.columns = [col.lower() for col in dataset.columns]
    dataset["year_of_release"] = pd.to_datetime(dataset["year_of_release"], format="%Y")
    return dataset
```

```python
dataset_path = data_path / "games.csv"
df = pd.read_csv(dataset_path)
df = prepare_dataset(df)
```

Let's analyse how global sales distributed in our dataset. To plot distributions using seaborn, we can use the `displot` method.

```python
sns.displot(data=df, x="global_sales")
plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_0.png",
    dpi=300,
    bbox_inches="tight",
)
```

As we can see the plot is not that informative. It is because of outliers in the dataset. Let's trye to logarithm the x-axis of the histogram and see the resulting graph.

```python
sns.displot(data=df, x="global_sales", log_scale=True)
plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_1.png",
    dpi=300,
    bbox_inches="tight",
)
```

But now let's get back to huge values which in the dataset and analyse them, trying to determine the range that we need.

```python
df.global_sales.describe()
```

Yes, as we predicted, while the mean and std are about 0.6, 1, the maximum value is about 82! For further analysis let's select the values which lay in the range until 95 percentile (or quantile 0.95).



```python
df_selected = df[df.global_sales <= df.global_sales.quantile(0.95)]
sns.displot(data=df_selected, x="global_sales")

plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_2.png",
    dpi=300,
    bbox_inches="tight",
)
```

While it's already looks understandable we can tune it a bit more with reducing number of bins:


```python
df_selected = df[df.global_sales <= df.global_sales.quantile(0.95)]
sns.displot(data=df_selected, x="global_sales", bins=10)

plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_3.png",
    dpi=300,
    bbox_inches="tight",
)
```

Great, we even can see a distribution. But besides determining the distribution type and its parameters, we can do nothing valuable here. Let's use the same techniques to compare distributions! For now let's try to compare `global_sales` of 2 different publishers: `Nintendo` and `Electronic Arts`, by filtering initial dataset and using `hue` parameter of the `displot` method.

```python
df_filtered = df[df["publisher"].isin(["Nintendo", "Electronic Arts"])]
df_filtered = df_filtered[df_filtered.global_sales <= df_filtered.global_sales.quantile(0.99)]

sns.displot(data=df_filtered, x="global_sales", hue="publisher", bins=20)

plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_4.png",
    dpi=300,
    bbox_inches="tight",
)
```

In the figure above we just plot two distributions above each other. From the first side it could look ok, but here is an invisible problem: when we are plotting the counts, visually we can see difference, while in reality there is not difference in actual distributions. To avoid it, we can use proportions (probabilities) instead:

```python
df_filtered = df[df["publisher"].isin(["Nintendo", "Electronic Arts"])]
df_filtered = df_filtered[df_filtered.global_sales <= df_filtered.global_sales.quantile(0.99)]

sns.displot(data=df_filtered, x="global_sales", hue="publisher", bins=20, stat="probability")

plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_5.png",
    dpi=300,
    bbox_inches="tight",
)
```

Often, it could be much easier to work with histograms if firstly preprocess the dataset. Let's focus on preprocessing our dataset and compare distributions of sales in different regions(`eu` and `jp`).

```python
needed_regions = ("eu_sales", "jp_sales")
df_tmp = df[[col for col in df.columns if "sales" in col and col in needed_regions]]
df_melted = df_tmp.melt(var_name="region", value_name="sales")
df_melted["region"] = df_melted["region"].apply(lambda x: x.split("_")[0])

# Quantile filtering
thresholds = df_melted.groupby("region")["sales"].quantile(0.95)
result_df = pd.DataFrame()

for region, threshold in thresholds.items():
    filtered_df = df_melted[(df_melted["region"] == region) & (df_melted["sales"] < threshold)]
    result_df = pd.concat([result_df, filtered_df])


sns.histplot(
    data=result_df,
    x="sales",
    hue="region",
    bins=10,
    element="step",
    stat="density",
    common_norm=False,
)

plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_6.png",
    dpi=300,
    bbox_inches="tight",
)
```

Also, as we discussed in the first week, distributions could be excellent addition to another visualization techniques to explain the data from another point of view

```python
col1, col2 = "user_score", "critic_score"
sns.jointplot(
    data=df,
    x=col1,
    y=col2,
    kind="reg",
    joint_kws={
        "scatter_kws": {"alpha": 0.1},
        "line_kws": {"color": "firebrick"},
    },
)

plt.savefig(
    figures_path / "step_by_step_tutorial" / "step_7.png",
    dpi=300,
    bbox_inches="tight",
)
```
