```python
import pandas as pd
```

```python
outlays = pd.read_csv('budget-outlays.csv', index_col=0).squeeze()
print(outlays.sum())
outlays
```

```python
outlays_threshold = 150000

outlays_others = outlays[outlays <= outlays_threshold]
outlays_others_value = outlays_others.sum()
outlays_others_value
```

```python
outlays_grouped = outlays[outlays > outlays_threshold].sort_values(ascending=False)
outlays_grouped['Others'] = outlays_others_value

outlays_grouped_normalized = (outlays_grouped / outlays_grouped.sum() * 100).round(3)
outlays_grouped_normalized.agg([lambda x: x, 'cumsum'])
```

```python
receipts = pd.read_csv('budget-receipts.csv', index_col=0).squeeze()
print(receipts.sum())
print(receipts.sum() / outlays.sum())
receipts
```

```python
receipts_normalized = ((receipts / receipts.sum() * 100) * receipts.sum() / outlays.sum()).round(2).sort_values(ascending=False)
receipts_normalized.agg([lambda x: x, 'cumsum'])
```
