```python
import pandas as pd
from pandas.core.dtypes.cast import convert_dtypes
```

# Preprocessing

```python
sp500_assets = pd.read_csv('sp500-assets.csv', index_col=0, header=[0, 1])

sp500_assets = sp500_assets.stack(level=1, future_stack=True).reset_index(level=1)
sp500_assets = sp500_assets.rename(columns={'level_1': 'Ticker'})

sp500_assets.index = pd.to_datetime(sp500_assets.index)

sp500_assets
```

```python
sp500_assets_close_stats = sp500_assets.groupby(pd.Grouper(freq='D')).apply(lambda group: group['Close'].notna().sum())
sp500_assets_close_stats[sp500_assets_close_stats == 503]
```

```python
date = pd.to_datetime('2022-12-29')

sp500_shares_subset_after = sp500_assets[sp500_assets.index == date]
sp500_shares_subset_after.to_csv(f'sp500-shares-{date.date()}.csv', index=False)

sp500_shares_subset_before = sp500_assets[sp500_assets.index == date - pd.Timedelta(days=1)]
sp500_shares_subset_before.to_csv(f'sp500-shares-{(date - pd.Timedelta(days=1)).date()}.csv', index=False)
```
