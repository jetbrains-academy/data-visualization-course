```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

root_path = Path().cwd().parent
data_path = root_path / "common" / "resources" / "datasets"
figures_path = Path().resolve() / "figures"
```

```python
x1 = np.arange(0, 100)
y1 = x1

x2 = np.arange(100, 125)
y2 = -2 * x2 + 300

x3 = np.arange(125, 200)
y3 = 1.5 * x3 - 140

x4 = np.arange(200, 280)
y4 = -0.5 * x4 + 260

x5 = np.arange(280, 300)
y5 = 5 * x5 - 1280

x = np.concatenate([x1, x2, x3, x4, x5])
y = np.concatenate([y1, y2, y3, y4, y5])

y = savgol_filter(y, window_length=100, polyorder=3)
y_noise = np.random.normal(0, 8, len(y))
y += y_noise

fig, ax = plt.subplots()
ax.plot(x[:250], y[:250])
ax.set_xticks([])
ax.set_yticks([])
ax.spines[["top", "right", "bottom", "left"]].set_visible(False)
fig.savefig(figures_path / "theory_trend_before.png", dpi=300, bbox_inches="tight")

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticks([])
ax.set_yticks([])
ax.spines[["top", "right", "bottom", "left"]].set_visible(False)
fig.savefig(figures_path / "theory_trend_after.png", dpi=300, bbox_inches="tight")

plt.show()
```
